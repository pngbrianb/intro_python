## What Is Python?

- Created by Guido Van Rossum in 1991
    - Despite logo, named after *Monty Python* comedy troupe
- Emphasizes **productivity** and **code readability**
- **Easy** to pick up and learn

````python
# An example of readable code:

a = 3
b = 200

def easy_to_read(number1,number2):
    if number1 < number2:
        print("This is easy to read!")
        
easy_to_read(a,b)
````

### Syntax

The syntax of a programming language is the set of rules that define the combinations of symbols that are considered to be correctly structured programs in that language


Just like our spoken languages have structures like paragraphs and sentences, programming languages have code blocks and statements.

### Variables

Variables are symbolic names that store a specific piece of information. Variables come in different types in order to hold different kinds of information.

For example: `r = 3`  
- Defines a variable: named “r”
- This holds a numerical integer: 3  

Other types of variables: 
- float (3.14)
- string (“Hello”)

### Control Structures

A block of programming that analyses variables and chooses a direction in which to go based on given parameters is a control structure.

The term flow control details the direction the program takes (how program logic “flows”). It determines how a computer will respond when given certain conditions and parameters. Some typical structures include:
- **If** statements

````python
# More quick examples

if num1 < num2:
    print('num1 is smaller!')
elif num1 == num2:
    print('num1 equals num2')
else:
    print('num1 is not smaller!')
````

- **For** loops

````python
for letter in 'this word':
    print(letter.upper())
````

- **Functions**

````python
def print_letter(a_string):
    
    for letter in a_string:

        print(letter.upper())
        
print_letter('nailed it!')
````

### Functions

- Groups of instructions repeat / create abstractions of common tasks. 
- Divide our code into useful blocks
- Provide order, making the code more readable and reusable
- `def name(input1, input2, …) :` (note the colon!)
- Definition saves instructions only - no execution!

**Optional, but good practice for published code:**
- First line is “Document String” describing how function works

````python
# an example function:
def square(number=2):
    """
    A docstring is a string literal that occurs 
    as the first statement in a module, function, 
    class, or method definition.
    
    Good practice for code readability, and some
    programs display docstrings in helpful ways
    """
    
    return number ** 2
````

````python
# now let's call our function with an input
````

````python
# and with the default
````

````python
# any function can only return ONE result
def sum_combinations(a,b,c):
    return a+b
    return b+c # doesn't execute
    return a+c # doesn't execute
````

````python
sum_combinations(2,4,8)
# returns only 2+4
````

A function can only `return` one result, so that you can use that result as an input for other functions.

(In our above example, you could `return` a single list with three different values)

````python
# Functions can be run off of the result of other functions:

square(square(3))
````
